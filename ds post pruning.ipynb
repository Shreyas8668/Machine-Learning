```python
from sklearn.datasets import  load_iris
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn import tree
```

```python
import matplotlib.pyplot as plt 
```

```python
iris =load_iris()
```

```python
iris
```

```python
X = iris.data
y=iris.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
clf=DecisionTreeClassifier(random_state=42)
clf.fit(X_train,y_train)
```

```python
print("Accuracy before pruning:", clf.score(X_test, y_test))
```

```python
path = clf.cost_complexity_pruning_path(X_train, y_train)
ccp_alphas = path.ccp_alphas       # candidate alphas
impurities = path.impurities
```

```python
clfs = []
for ccp_alpha in ccp_alphas:
    model = DecisionTreeClassifier(random_state=42, ccp_alpha=ccp_alpha)
    model.fit(X_train, y_train)
    clfs.append(model)
```

```python
train_scores = [model.score(X_train, y_train) for model in clfs]
test_scores = [model.score(X_test, y_test) for model in clfs]
```

```python
plt.figure(figsize=(10, 6))
plt.plot(ccp_alphas, train_scores, marker='o', label="Train Accuracy")
plt.plot(ccp_alphas, test_scores, marker='o', label="Test Accuracy")
plt.xlabel("ccp_alpha")
plt.ylabel("Accuracy")
plt.title("Post-Pruning: Accuracy vs CCP Alpha")
plt.legend()
plt.show()
```

```python
best_alpha_index = test_scores.index(max(test_scores))
best_tree = clfs[best_alpha_index]
```

```python
print("Best ccp_alpha:", ccp_alphas[best_alpha_index])
print("Accuracy after pruning:", best_tree.score(X_test, y_test))
```
